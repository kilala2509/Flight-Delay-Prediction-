Read csv files

```python
d1=pd.read_csv('/content/drive/MyDrive/ML Project/Detailed_Statistics_Departures-CHICAGOO.csv')
d2=pd.read_csv('/content/drive/MyDrive/ML Project/Detailed_Statistics_Departures-DEN.csv')
d3=pd.read_csv('/content/drive/MyDrive/ML Project/Detailed_Statistics_Departures-EWR.csv')
d4=pd.read_csv('/content/drive/MyDrive/ML Project/Detailed_Statistics_Departures-IAD.csv')
d5=pd.read_csv('/content/drive/MyDrive/ML Project/Detailed_Statistics_Arrivals-2.csv')
```

Remove all null values

```python
d11=d1.dropna()
```

```python
d22=d2.dropna()
```

```python
d33=d3.dropna()
```

```python
d44=d4.dropna()
```

```python
d55=d5.dropna()
```

Combine the departure dataframes using concatenation

```python
flightsinfo = pd.concat([d11,d22,d33,d44], axis=0)
```

```python
flightsinfo
```

Filter rows for flights going to Syracuse Airport

```python
flightsinfo = flightsinfo[flightsinfo['Destination Airport'].str.contains('SYR')]
flightsinfo.head(20)
```

Merge the departure dataframe and arrival dataframe on Date and Flight number

```python
df=pd.merge(flightsinfo,d5,on=['Date (MM/DD/YYYY)','Flight Number'])
```

```python
df
```

```python
df.columns
```

Remove duplicate columns

```python
df = df.T.drop_duplicates().T
```

```python
df
```

```python
df.columns
```

Drop unnecessary columns

```python
df=df.drop(columns=['Scheduled Arrival Time', 'Actual Arrival Time','Carrier Code_x','Tail Number_x','Scheduled departure time',
       'Actual departure time'])
```

```python
df
```

Change the date column into seperate day month year columns

```python
df[["month", "day", "year"]] = df["Date (MM/DD/YYYY)"].str.split("/", expand = True)
```
