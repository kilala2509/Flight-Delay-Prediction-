Read all the csv files into dataframe

**FOR WASHINGTON**

```python
d1=pd.read_csv('/content/drive/MyDrive/ML Project/washington.csv')
```

```python
d1
```

split datetime column into day month year columns

```python
d1[["day", "month", "year"]] = d1["datetime"].str.split("/", expand = True)
```

drop unnecessary columns

```python
d1=d1.drop(columns=['datetime','stations', 'description', 'sunrise', 'sunset','conditions','preciptype'])
```

check for null values

```python
d1.isna().sum()
```

drop null values

```python
d1=d1.dropna()
```
