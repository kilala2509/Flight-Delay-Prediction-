import necessary libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
df1=pd.read_csv('/content/drive/MyDrive/ML Project/flightinfo-3.csv')
```

```python
df2=pd.read_csv('/content/drive/MyDrive/ML Project/WeatherDataF.csv')
```

```python
df1
```

```python
df2
```

Merge the data on Origin Airport and Date

```python
df=pd.merge(df1,df2,on=['Origin Airport','day','month','year'])
```

```python
df
```

```python
df.columns
```

Perform label encoding on the columns which have string datatype

```python
le=LabelEncoder()
```

```python
df['Origin Airport'] = le.fit_transform(df['Origin Airport'])
df['Destination Airport'] = le.fit_transform(df['Destination Airport'])
df['icon'] = le.fit_transform(df['icon'])
```

Convert into a csv file

```python
df.to_csv('Result.csv')
```
